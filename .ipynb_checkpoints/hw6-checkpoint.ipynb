```python
import numpy as np
from scipy.optimize import differential_evolution
```

**Задание**

При фиксированном seed=42 поразбирайтесь и поэкспериментируйте с параметрами алгоритма: strategy, maxiter, popsize, tol, mutation, recombination и updating.

Постройте графики количества итераций оптимизации функции ackley от значения параметра.

Допзадание
Найти множество параметров (strategy=..., maxiter=..., ..., updating=...), которые требуют минимального количества итераций.

```python
def ackley(x):
    arg1 = -0.2 * np.sqrt(0.5 * (x[0]**2 + x[1]**2))
    arg2 = 0.5 * (np.cos(2. * np.pi * x[0]) + np.cos(2. * np.pi * x[1]))
    return -20. * np.exp(arg1) - np.exp(arg2) + 20. + np.e

bounds = [(-10, 10), (-10, 10)]
result = differential_evolution(ackley, bounds, seed=42)
result
```

```python
def ackley_2(x, y):
    arg1 = -0.2 * np.sqrt(0.5 * (x**2 + y** 2))
    arg2 = 0.5 * (np.cos(2. * np.pi * x) + np.cos(2. * np.pi * y))
    return -20. * np.exp(arg1) - np.exp(arg2) + 20. + np.e
```

```python
def makeData():
    # Строим сетку в интервале от -10 до 10 с шагом 0.1 по обоим координатам
    x = np.arange(-10, 10, 0.1)
    y = np.arange(-10, 10, 0.1)

    # Создаем двумерную матрицу-сетку
    xgrid, ygrid = np.meshgrid(x, y)

    # В узлах рассчитываем значение функции
    zgrid = ackley_2(xgrid, ygrid)
    return xgrid, ygrid, zgrid
```

```python
import pylab
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

x, y, z = makeData()

fig = pylab.figure(figsize=[12, 9])
axes = Axes3D(fig)

axes.plot_surface(x, y, z, rstride=20, cstride=1)
pylab.show()
```

```python
# Постройте графики количества итераций оптимизации функции ackley от значения параметра.
# strategy, maxiter, popsize, tol, mutation, recombination и updating.
strategies = ['best1bin', 'best1exp', 'rand1exp', 'randtobest1exp','currenttobest1exp', 'best2exp',
            'rand2exp', 'randtobest1bin', 'currenttobest1bin', 'best2bin', 'rand2bin', 'rand1bin']
#maxiter: (maxiter + 1) * popsize * len(x)
#mutation: If specified as a float it should be in the range [0, 2]
#updating : {'immediate', 'deferred'}
```

```python
i = []
for strategy in strategies:
    result = differential_evolution(ackley, bounds, seed=42, strategy=strategy)
    i.append(result.nit)
```

```python
plt.figure(figsize=[15, 4])
plt.plot(i)
plt.title('number of iterations depending on strategy')
plt.xlabel('startegy: from best1bin to rand1bin')
plt.ylabel('number of iterations')
plt.show()
```

Вывод: лучшей стратегией все же является стратегия по умолчанию - best1bin. Именно она выдает минимальное количество итераций - 89. Будем выбирать именно ее.

```python
result = differential_evolution(ackley, bounds, seed=42, maxiter=100)
result
```

Вывод: с выбранной стратегией параметр maxiter не влияет на кол-во итераций. он влияет на то, получен успех или нет.

```python
popsize_result = []
x_grid = []
for popsize in range(100):
    result = differential_evolution(ackley, bounds, seed=42, popsize=popsize)
    if result.success  == True:
        popsize_result.append(result.nit)
        x_grid.append(popsize)
```

```python
plt.figure(figsize=[15, 4])
plt.plot(x_grid, popsize_result)
plt.title('number of iterations depending on popsize')
plt.xlabel('popsize')
plt.ylabel('number of iterations')
plt.show()
```
